```python
import pandas as pd
columns_to_set_mode_to_zero = ['V16', 'V17', 'V18', 'V19', 'V20', 'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28']
mode_values = creditcard[columns_to_set_mode_to_zero].mode().iloc[0]
creditcard[columns_to_set_mode_to_zero] = creditcard[columns_to_set_mode_to_zero].fillna(0)
print(creditcard.head())
```

```python
creditcard.isnull().sum()
```

```python
print(creditcard['Amount'].mode())
```

```python
print(creditcard['Amount'].mode()[0])
```

```python
creditcard['Amount'].fillna(creditcard['Amount'].mode()[0], inplace=True)
```

```python
print(creditcard['Class'].mode())
```

```python
print(creditcard['Class'].mode()[0])
```

```python
creditcard['Class'].fillna(creditcard['Class'].mode()[0], inplace=True)
```

```python
creditcard.isnull().sum()
```

```python
creditcard['Class'].value_counts()
```

```python
legit = creditcard[creditcard.Class == 0]
fraud = creditcard[creditcard.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comparing values for both transactions
creditcard.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=156)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

splitting data

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

logistic regression model

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

accuracy

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction= model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data:',test_data_accuracy)
```
